```python
import os
import pandas
import folium
from folium import IFrame
import random
```

# folium 모듈을 이용하면 쉽게 커스텀지도를 만들 수 있습니다. 
# html파일의 형태로 바로 만들어집니다.

```python
min_lat=-70
max_lat=110
min_lon=20
max_lon=340

mymap = folium.Map(location=[15.9613,206.3672], 
                                zoom_start=2,
                                tiles='cartodbdark_matter',
                                max_bounds=True,
                                min_lat=min_lat,
                                max_lat=max_lat,
                                min_lon=min_lon,
                                max_lon=max_lon,
                                max_zoom=4,
                                min_zoom=2)
mymap
```

# 지도에서 클릭한 지점의 x, y좌표를 알려줍니다.

```python
mymap.add_child(folium.LatLngPopup())
mymap
```

# 커스텀이미지마커를 만듭니다. 그리고 그 이미지마커를 클릭하면 ifrmae태그를 이용해 html파일을 마커에 추가시킵니다 

```python
news_where = [36.4033,127.8618]
news_msg = '일회용 수저 빼달랬는데 왜 자꾸 주나요'
news_img_link = 'plsnews/static/plsnews/images/spoons.jpg'
news_img = folium.features.CustomIcon(news_img_link, icon_size=(100,50))

news_link = 'http://www.segye.com/newsView/20191004509710?OutUrl=Zum'
news_atag ='<a href=' + news_link + ' target="_blank">see the article</a>' 

news_iframe = IFrame(news_atag, width=100, height=50)
news_popup = folium.Popup(news_iframe, max_width=2650, parse_html=True)

folium.Marker(news_where,
              tooltip=news_msg,
              icon=news_img, 
              popup=news_popup).add_to(mymap)

mymap
```

# image overlay를 이용해서 지도에 이미지를 겹칩니다

```python
min_lat=-70
max_lat=110
min_lon=20
max_lon=340

mymap = folium.Map(location=[15.9613,206.3672], 
                                zoom_start=2,
                                tiles='cartodbdark_matter',
                                max_bounds=True,
                                min_lat=min_lat,
                                max_lat=max_lat,
                                min_lon=min_lon,
                                max_lon=max_lon,
                                max_zoom=4,
                                min_zoom=2)

gyre_img_link = 'plsnews/static/plsnews/images/whirlpool.png'
gyre_img = folium.raster_layers.ImageOverlay(
    name='plastic island',
    image=gyre_img_link,
    bounds=[[2, 160], [50, 220]],
    opacity=0.4,
    interactive=True,
    cross_origin=False,
    zindex=1,
)
folium.Popup('i am the gyre').add_to(gyre_img)

gyre_img.add_to(mymap)
folium.LayerControl().add_to(mymap)

mymap
```

# iframe에 이미지를 넣을 때는 인터넷상의 이미지를 가져오는게 렌더링이 빠르다 
# 이미지 찾는곳: https://commons.wikimedia.org/wiki/Main_Page

```python
from folium import IFrame

resolution, width, height = 75, 7, 3

html = '<img src="https://upload.wikimedia.org/wikipedia/commons/f/f4/Mercator_projection_SW.jpg">'
iframe = IFrame(html, width=(width*resolution)+20, height=(height*resolution)+20)
popup = folium.Popup(iframe, max_width=2650)

icon = folium.Icon(color="red", icon="ok")
marker = folium.Marker(location=[-21.2520, 136.0329], popup=popup, icon=icon)
marker.add_to(mymap)

mymap
```
